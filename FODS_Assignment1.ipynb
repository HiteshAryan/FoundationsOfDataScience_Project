```python
import numpy as np
from scipy.stats import beta
o = 48
z = 160 - o
ones = np.ones(o, dtype = 'int')
print(ones)
```

```python
zeros = np.zeros(z, dtype = 'int')
print(zeros)
```

```python
r = np.concatenate((ones, zeros))
print(r)
```

```python
import random
random.shuffle(r)
print(r)
```

```python
import matplotlib.pyplot as plt 
from scipy.special import gamma
a = 2
b = 3
m = o
l = z
mu = np.arange(0, 1, 0.01)
pr = beta.pdf()
plt.plot(mu, pr)
plt.show()
```

```python
import scipy.integrate
y = lambda mu: (gamma(a+b)/(gamma(a)*gamma(b)))*(mu**(a-1)*(1-mu)**(b-1))
scipy.integrate.quad(y, 0, 1)
```

```python
u = np.arange(0, 1, 0.01)
post = (gamma(m+a+l+b)/(gamma(m+a)*gamma(l+b)))*(u**(m+a-1)*(1-u)**(l+b-1))
plt.plot(u, post)
plt.show()
```
